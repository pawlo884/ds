# Eksplorujemy `pycaret` 

## Czym jest `pycaret`?

* `pycaret` to biblioteka Pythonowa, która pozwala na szybkie i łatwe tworzenie modeli uczenia maszynowego.
* `pycaret` pozwala na automatyzację wielu kroków w procesie tworzenia modeli uczenia maszynowego, takich jak:
    * przygotowanie danych
    * tworzenie modeli
    * porównywanie modeli
    * interpretacja modeli
    * tworzenie raportów
    * oraz wiele innych
* `pycaret` będzie przez nas używany do trenowania własnych modeli uczenia maszynowego.

## Instalujemy wymagane biblioteki (conda wymagana)

- otwórz terminal
- uruchom `conda activate od_zera_do_ai`
- uruchom `conda install -y 'pycaret[full]'`

## Klasy modeli - Klasyfikacja

Klasyfikacja to zadanie polegające na przyporządkowaniu obiektu do jednej z kategorii. 


## Klasy modeli - Regresja

Regresja to zadanie polegające na przewidywaniu wartości numerycznej na podstawie innych zmiennych.



## Klasy modeli - Klastrowanie

Klastrowanie to zadanie polegające na grupowaniu obiektów na podstawie ich podobieństwa.


## Klasy modeli - Wykrywanie anomalii (Anomaly Detection)

Wykrywanie anomalii to zadanie polegające na identyfikowaniu obiektów, które znacznie różnią się od reszty zbioru danych.
