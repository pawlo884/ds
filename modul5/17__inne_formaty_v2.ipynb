# Inne formaty

Oprócz już wspomnianych formatów, istnieje wiele innych, które mogą być przydatne w różnych sytuacjach. Poniżej znajdziesz kilka przykładów.

## Format Parquet

![](https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/Apache_Parquet_logo.svg/800px-Apache_Parquet_logo.svg.png?20210417030530)

* Parquet to kolumnowy format przechowywania danych zoptymalizowany do przetwarzania danych na dużą skalę. 
* Jest zaprojektowany do efektywnego obsługiwania złożonych zagnieżdżonych struktur danych
* Oferuje wysokie współczynniki kompresji oraz szybszą wydajność zapytań w porównaniu do formatów wierszowych, takich jak CSV.

````python
# jak zapisać dane do parquet
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df.to_parquet('data.parquet')
````

````python
# jak wczytać dane z parquet
df = pd.read_parquet('users.parquet', engine='pyarrow')
df
````

## Format ORC

![](https://upload.wikimedia.org/wikipedia/commons/thumb/e/e7/Apache_Orc_logo.svg/450px-Apache_Orc_logo.svg.png?20210417024622)

* ORC to kolejny kolumnowy format przechowywania danych
* Zaprojektowany w celu zapewnienia wysokiej efektywności przechowywania, odczytu i kompresji. 

````python
# jak zapisać dane do ORC
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df.to_orc('data.orc', engine='pyarrow')
````

````python
# jak wczytać dane z ORC
df = pd.read_orc('users.orc')
df
````
