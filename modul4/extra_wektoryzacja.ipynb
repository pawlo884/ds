# O wektoryzacji w Pandas

* W Pandas mamy do dyspozycji wiele metod, które pozwalają na wykonywanie operacji na całych kolumnach jednocześnie. 
* Dzięki temu możemy uniknąć pętli `for` i zwiększyć wydajność naszego kodu. 

```python
import pandas as pd
```

```python
df = pd.DataFrame([
    {"cena": 1, "nazwa": "A"},
    {"cena": 6, "nazwa": "B"},
    {"cena": 17, "nazwa": "C"},
    {"cena": 61, "nazwa": "D"},
])
df
```

```python
df["cena"] + 5
```

## jeżeli chcesz żeby nowe wartości były przypisane do kolumny "cena"

```python
df["cena"] = df["cena"] + 5
df
```

## jeżeli dodajesz nową kolumnę używając np. liczby wówczas też mamy do czynienia z wektoryzacją

```python
df["ilość"] = 45
df
```

## Jak i kiedy działa wektoryzacja? Praktyczny przykład

```python
from datetime import datetime

now = datetime.now()
hire_date = [datetime(2024, 1, 13), datetime(2024, 2, 13), datetime(2024, 3, 13)]

# poniższa linijka wywoła błąd: `TypeError: unsupported operand type(s) for -: 'datetime.datetime' and 'list'`
# bo nie można odejmować listy dat od daty
now - hire_date  # <== to nie zadziała
```

```python
# ale co się stanie jeśli załadujemy listę hire_date do DataFrame?
df = pd.DataFrame({"hire date": hire_date})
df
```

```python
# wówczas ta operacja zadziała, bo Pandas zrobi wektoryzację `now`
now - df["hire date"]
```
