```python
import pandas as pd
```

# Jak łączyć ze sobą DataFrame'y?

## Sklejanie danych jeden pod drugim - konkatencja!



```python
df0 = pd.DataFrame([
    {'imię': 'Ada', 'b': 3},
    {'imię': 'Paweł', 'b': 6},
    {'imię': 'Kasia', 'b': 9},
    {'imię': 'Jacek', 'b': 12},
])
df1 = pd.DataFrame([
    {'imię': 'Basia', 'b': 15},
    {'imię': 'Adrian', 'b': 17},
    {'imię': 'Julia', 'b': 21},
])

df = pd.concat([df0, df1], ignore_index=True)
df
```

## Łączenie danych po wspólnym kluczu / kolumnie - INNER JOIN



```python
df0 = pd.DataFrame([
    {'imię': 'Ada', 'b': 3},
    {'imię': 'Paweł', 'b': 6},
    {'imię': 'Kasia', 'b': 9},
    {'imię': 'Jacek', 'b': 12},
])
df1 = pd.DataFrame([
    {'imię': 'Ada', 'c': 100},
    {'imię': 'Kasia', 'c': 200},
    {'imię': 'Romek', 'c': 300},
])

df = df0.merge(df1, on='imię', how='inner')
df
```

## Łączenie danych po wspólnym kluczu / kolumnie, chroniąc lewy DataFrame - LEFT JOIN



```python
df0 = pd.DataFrame([
    {'imię': 'Ada', 'b': 3},
    {'imię': 'Paweł', 'b': 6},
    {'imię': 'Kasia', 'b': 9},
    {'imię': 'Jacek', 'b': 12},
])
df1 = pd.DataFrame([
    {'imię': 'Ada', 'c': 100},
    {'imię': 'Kasia', 'c': 200},
    {'imię': 'Romek', 'c': 300},
])

df = df0.merge(df1, on='imię', how='left')
df
```

## Łączenie danych po wspólnym kluczu / kolumnie, chroniąc prawy DataFrame - RIGHT JOIN


```python
df0 = pd.DataFrame([
    {'imię': 'Ada', 'b': 3},
    {'imię': 'Paweł', 'b': 6},
    {'imię': 'Kasia', 'b': 9},
    {'imię': 'Jacek', 'b': 12},
])
df1 = pd.DataFrame([
    {'imię': 'Ada', 'c': 100},
    {'imię': 'Kasia', 'c': 200},
    {'imię': 'Romek', 'c': 300},
])

df = df0.merge(df1, on='imię', how='right')
df
```

## Łączenie danych po wspólnym kluczu / kolumnie, chroniąc obydwa DataFrame'y - OUTER JOIN


```python
df0 = pd.DataFrame([
    {'imię': 'Ada', 'b': 3},
    {'imię': 'Paweł', 'b': 6},
    {'imię': 'Kasia', 'b': 9},
    {'imię': 'Jacek', 'b': 12},
])
df1 = pd.DataFrame([
    {'imię': 'Ada', 'c': 100},
    {'imię': 'Kasia', 'c': 200},
    {'imię': 'Romek', 'c': 300},
])

df = df0.merge(df1, on='imię', how='outer')
df
```
