# Eksploracja DataFrame

Zanim zaczniemy musimy przygotować DataFrame, na którym będziemy pracować. W tym celu wygenerujemy losowe dane. 

```python
# lista, w której będziemy przechowywać losowe dane
random_data = []

# stworzymy 100 rekordów
for i in range(100):
    random_data.append({
        # random.choice wybiera losowy element z listy, w tym przypadku losujemy imię
        "name": random.choice(["Adam", "Ada", "Basia", "Jan", "Kasia"]),

        # random.randint losuje liczbę całkowitą z podanego zakres, w tym przypadku wiek od 18 do 80
        "age": random.randint(18, 80),

        # random.randint losuje liczbę całkowitą z podanego zakres, w tym przypadku wzrost od 150 do 200
        "height": random.randint(150, 200),

        # random.choice wybiera losowy element z listy, w tym przypadku losujemy miasto
        "city": random.choice(["Wrocław", "Warszawa", "Szczecin", "Kraków", "Praga"])
    })

# tworzymy DataFrame z losowymi danymi
df = pd.DataFrame(random_data)
```
