# Jak wizualizować DataFrame'y?

## Histogramy



```python
df = pd.DataFrame([
    {"name": "Alice", "age": 21, "height": 155},
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 31, "height": 190},
])

df.hist()
```

```python
df = pd.DataFrame([
    {"name": "Alice", "age": 21, "height": 155},
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 31, "height": 190},

])

df['age'].hist(bins=3)
```

# Wykres kołowy - pie chart



```python
df = pd.DataFrame([
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 23, "height": 190},

])

df.plot(kind='pie', y='age', labels=df['name'], legend=False)
```

# Wykres słupkowy - barplot



```python
df = pd.DataFrame([
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 23, "height": 190},
])

df.plot(kind='bar', x='name', y=['age', 'height'])
```

# Wykres punktowy - scatter plot



```python
df = pd.DataFrame([
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 23, "height": 190},
])

df.plot(kind='scatter', x="height", y='age')
```

# Wykres pudełkowy - boxplot



```python
df = pd.DataFrame([
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 23, "height": 190},
])

df.plot(kind='box')
```

## Jak zwizualizować korelację między zmiennymi?

```python
df = pd.DataFrame([
    {"name": "Alice", "age": 24, "height": 160},
    {"name": "Bob", "age": 30, "height": 155},
    {"name": "Charlie", "age": 20, "height": 165},
    {"name": "Andrzej", "age": 21, "height": 192},
    {"name": "Adam", "age": 23, "height": 190},
])

df[["age", "height"]].corr()
```
