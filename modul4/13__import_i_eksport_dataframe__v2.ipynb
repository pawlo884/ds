## Zapisywanie DataFrame do pliku CSV



```python
df = pd.DataFrame([
    {'month': 'January', 'revenue': 100},
    {'month': 'February', 'revenue': 200},
    {'month': 'March', 'revenue': 300},
    {'month': 'April', 'revenue': 400},
])
df.to_csv('revenue.csv', index=False)
```

## Zapisywanie DataFrame do pliku Excel




```python
df = pd.DataFrame([
    {'month': 'January', 'revenue': 100},
    {'month': 'February', 'revenue': 200},
    {'month': 'March', 'revenue': 300},
    {'month': 'April', 'revenue': 400},
])
df.to_excel('revenue.xlsx', index=False)
```

## Zapisywanie DataFrame do pliku Excel - zapisywanie do konkretnego arkusza


```python
df = pd.DataFrame([
    {'month': 'January', 'revenue': 100},
    {'month': 'February', 'revenue': 200},
    {'month': 'March', 'revenue': 300},
    {'month': 'April', 'revenue': 400},
])
df.to_excel('revenue.xlsx', index=False, sheet_name='revenue')
```
