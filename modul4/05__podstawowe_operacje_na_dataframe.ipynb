# Podstawowe operacja na DataFrame

Zanim zaczniemy musimy przygotować DataFrame, na którym będziemy pracować. 

```python
df = pd.DataFrame([
    {"name": "Adam", "age": 25, "weight": 70, "height": 180},
    {"name": "Alicja", "age": 30, "weight": 60, "height": 170},
    {"name": "Bartek", "age": 35, "weight": 100, "height": 160},
])
df
```

## Jak uzyskać dostęp do jednej kolumny w DataFrame?

```python
df["age"]
```

## Jak dodać nową kolumnę ze stałą wartością?


```python
df["country"] = "Poland"

df
```

## Jak dodać nową kolumnę z wartościami obliczonymi na podstawie innej kolumny?

```python
df["height_in_m"] = df["height"] / 100

df
```

## Jak dodać nową kolumnę z wartościami obliczonymi na podstawie dwóch innych kolumn?

```python
df["bmi"] = df["weight"] / (df["height"] / 100) ** 2

df
```

## Jak dodać kolumnę używając bardziej wyszukany obliczeń?

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "underweight"
    elif bmi < 25:
        return "normal"
    elif bmi < 30:
        return "overweight"
    else:
        return "obese"


df["bmi_category"] = df["bmi"].apply(bmi_category)

df
```

## Jak usunąć kolumnę?



```python
df.drop(columns=["height_in_m"], inplace=True)

df
```

## Jak nadpisać kolumnę w Pandas DataFrame?



```python
df["age"] = df["age"] + 1

df
```

## Jak zmienić nazwę kolumny?



```python
df = df.rename(columns={"name": "first_name"})

df
```

## Jak zmienić kolejność kolumn w DataFrame?

```python
# pokażmy oryginalny DataFrame
df
```

```python
df[['age', 'bmi', 'bmi_category', 'country', 'first_name', 'height', 'weight']]
```

## Jak posortować DataFrame po kolumnie - rosnąco?

```python
# pokażmy oryginalny DataFrame
df
```

```python
df.sort_values(by="height")
```

## Jak posortować DataFrame po kolumnie - malejąco?

```python
# pokażmy oryginalny DataFrame
df
```

```python
df.sort_values(by="age", ascending=False)
```

## WAŻNE! Jak przekształcać DataFrame w miejscu?

* Większość operacji które poznaliśmy lub poznamy produkują nowy obiekt DataFrame. 
* Jeżeli chcemy, żeby te transformacje od razu nadpisały oryginalny DataFrame, musimy użyć parametru `inplace=True`.

```python
# pokażmy oryginalny DataFrame
df
```

```python
df.sort_values(by="age", ascending=False, inplace=True)
```
