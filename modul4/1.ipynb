```python
import pandas as pd
import random
```

```python
random_data = []

for i in range(100):
    random_data.append({
        "name": random.choice(["Ania", "Bartek", "Czesław", "Damian", "Eryk", "Filip", "Grzegorz", "Hubert", "Ignacy", "Jan"]),
        "age": random.randint(18, 80),
        "height": random.randint(150, 200),
        "city": random.choice(["Warszawa", "Kraków", "Wrocław", "Poznań", "Gdańsk", "Szczecin", "Katowice", "Lublin", "Białystok", "Łódź"]),
    })

df = pd.DataFrame(random_data)
```
