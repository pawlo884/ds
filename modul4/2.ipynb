```python
import pandas as pd
```

```python
df = pd.DataFrame([
    {"name": "Ania", "age": 20, "height": 160, "weight": 60},
    {"name": "Bartek", "age": 25, "height": 180, "weight": 80},
    {"name": "Czesław", "age": 30, "height": 170, "weight": 70},
    {"name": "Damian", "age": 35, "height": 175, "weight": 75},
    {"name": "Eryk", "age": 40, "height": 180, "weight": 80},
    {"name": "Filip", "age": 45, "height": 185, "weight": 85},
    {"name": "Grzegorz", "age": 50, "height": 190, "weight": 90},
    {"name": "Hubert", "age": 55, "height": 195, "weight": 95},
])
df
```

```python
df["age"]
```

```python
df["country"] = "Polska"
df
```

```python
df["height_in_m"] = df["height"] / 100
df
```

```python
df["bmi"] = df["weight"] / ((df["height"] /100) ** 2)
df
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "niedowaga"
    elif bmi < 25:
        return "waga prawidłowa"
    elif bmi < 30:
        return "nadwaga"
    else:
        return "otyłość"

df["bmi_category"] = df["bmi"].apply(bmi_category)
df
```

```python
df.drop(columns=["height_in_m"], inplace=True)
df
```

```python
df['age'] = df['age'] + 1
df
```

```python
df = df.rename(columns={"name": "first_name"})
df
```
