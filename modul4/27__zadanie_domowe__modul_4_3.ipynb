* Twoim zadaniem jest wykonanie obróbki danych zawartych w załączonym pliku CSV.
* Pozbądź się wszystkich rzędów które zawierają duplikaty
* Rozbij kolumnę `'Name'` na dwie kolumny `'Imie'` i `'Nazwisko'`
* Usuń kolumnę `'Name'`
* Napraw rzędy które zawierają brakuące dane
* Zmień nazwy kolumn:
    - `'Age'` -> `'Wiek'`
    - `'Height'` -> `'Wzrost'`
    - `'Weight'` -> `'Waga'`
    - `'Address'` -> `'Adres'`
      
* Dodaj kolumnę 'BMI' obliczoną jako `Waga / (Wzrost / 100) ^ 2`
* Posortuj DataFrame po kolumnie 'BMI' malejąco
* Zapisz zmodyfikowany DataFrame do pliku CSV
* Wykorzystaj nowo nabytą wiedzę z biblioteki Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('27__dane.csv', sep=';')

df
```

```python
# duplikaty
df.drop_duplicates(inplace=True)
df
```

```python
df[["Imię", 'Nazwisko']] = df['Name'].str.split(' ', expand=True)
df.drop(columns=['Name'], inplace=True)
df
```

```python
df["Age"] = df['Age'].fillna(df['Age'].mean()).round(0)
df['Height'] = df['Height'].fillna(df['Height'].mean()).round(0)
df['Weight'] = df['Weight'].fillna(df['Weight'].mean()).round(0)


def random_address(x):
    if pd.isnull(x):
        return df['Address'].dropna().sample(1).values[0]
    return x

def random_name(x):
    if pd.isnull(x):
        return df['Imię'].dropna().sample(1).values[0]
    return x

def random_surname(x):
    if pd.isnull(x):
        return df['Nazwisko'].dropna().sample(1).values[0]
    return x


df['Address'] = df['Address'].apply(random_address)
df['Imię'] = df['Imię'].apply(random_name)
df['Nazwisko'] = df['Nazwisko'].apply(random_surname)

df
```

```python
df.columns = ['Wiek', 'Adres', 'Wzrost', 'Waga', 'Imię', "Nazwisko"]
df
```

```python
df["BMI"] = (df["Waga"] / ((df["Wzrost"]/100) ** 2)).round(2)
df.sort_values(by="BMI", ascending=False)
```

```python
df.to_csv('zmodyfikowany_DataFrame.csv', index=False)
```
