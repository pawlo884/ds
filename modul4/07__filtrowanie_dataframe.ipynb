# Filtrowanie danych w DataFrame

Zanim zaczniemy musimy przygotować DataFrame, na którym będziemy pracować. 

```python
df = pd.DataFrame(
    [
        {"name": "Adam", "age": 25, "weight": 70, "height": 180},
        {"name": "Alicja", "age": 30, "weight": 60, "height": 170},
        {"name": "Bartek", "age": 35, "weight": 100, "height": 160},
        {"name": "Celina", "age": 42, "weight": 55, "height": 150},
    ],
    index=["pracownik HR 0", "pracownik HR 1", "pracownik IT 0", "pracownik IT 1"],
)

df
```

## Jak wybrać tylko kolumny `name` i `age`?



```python
df[["name", "age"]]
```

## Jak przefiltrować DataFrame ale tak żeby móc go wykorzystywać dalej, bez zmiany oryginalnego DataFrame?

```python
# żebyśmy stworzyć kopię dataframe'u, a nie widok na niego, musimy użyć metody .copy
name_and_age_df = df[["name", "age"]].copy()

# teraz możemy zmieniać wartości w kopii, a oryginalny dataframe nie zmieni się
name_and_age_df["age"] = name_and_age_df["age"] + 1

name_and_age_df
```

```python
# faktycznie zmiany nie wpłynęły na oryginalny dataframe
df
```

## Jak wybrać osoby tylko z wagą powyżej 65 kg?



```python
df[df["weight"] > 65]
```

## Jak wybrać osoby tylko z wagą powyżej 65 kg i wzrostem poniżej 170 cm?

```python
df[(df["weight"] > 65) & (df["height"] < 170)]
```

## Jak wybrać osoby tylko z wagą powyżej 65 kg lub wzrostem poniżej 170 cm?

```python
df[(df["weight"] > 65) | (df["height"] < 170)]
```
