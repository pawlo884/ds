# Podstawy uczenia maszynowego

## Modele uczenia maszynowego można podzielić na dwie główne kategorie **nadzorowane** i **nienadzorowane** 

* Modele **nadzorowane** wymagają danych wejściowych wraz z **etykietami / wartościami**, które są wykorzystywane do trenowania modelu. 
* Modele **nienadzorowane** nie wymagają etykiet i są wykorzystywane do grupowania danych lub redukcji wymiarów.


## Model **nadzorowany** - klasyfikacja

Klasyfikacja to zadanie polegające na przypisaniu obiektu do jednej z kategorii na podstawie jego cech.

## Model **nadzorowany** - regresja

Regresja to zadanie polegające na przewidywaniu wartości ciągłej na podstawie cech obiektu.

## Model **nienadzorowany** - klastrowanie / grupowanie

Klastrowanie to zadanie polegające na grupowaniu obiektów na podstawie ich podobieństwa.
