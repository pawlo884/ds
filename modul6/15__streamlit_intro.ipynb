# Tworzenie interaktywnych aplikacji - Streamlit - wprowadzenie

Biblioteka `Streamlit` pozwala na tworzenie interaktywnych aplikacji webowych w języku Python. 

Linki:
* Strona biblioteki `Streamlit`: [https://streamlit.io/](https://streamlit.io/)
* Dokumentacja biblioteki `Streamlit`: [https://docs.streamlit.io/](https://docs.streamlit.io/)
* Dokumentacja  - przegląd komponentów: [https://docs.streamlit.io/develop/api-reference](https://docs.streamlit.io/develop/api-reference)




## Jak zainstalować bibliotekę `Streamlit`?

Otwórz `terminal` lub `anaconda prompt` i wykonaj poniższe komendy.

Używając `conda`:
```bash
conda activate od_zera_do_ai
conda install -y streamlit
```

Lub używając `pip`:
```bash
pip install streamlit
```

Dodatkowo w tym module będziemy korzystać z biblioteki `openai` oraz `python-dotenv`, więc warto zainstalować je wcześniej:

```bash
pip install openai
pip install python-dotenv
```

## Jak będziemy pracować z biblioteką `Streamlit`?

### Tworzenie struktur

1. idź do folderu gdzie trzymasz wszystki pliki z tego modułu
2. utwórz folder dla nowej aplikacji (na filmach będę pokazywał jak je nazywać)
3. otwórz folder w `VS Code`
4. utwórz pusty plik `app.py`

### Uruchamianie aplikacji

W `VS Code` otwórz terminal i wpisz:
 
```bash
streamlit run app.py
```

jak nie chcesz żeby od razu otwierało przeglądarkę to:
```bash
streamlit run --server.headless=true app.py
```
