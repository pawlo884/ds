# Typy danych w Pythonie

## Witajcie w ogrodzie zoologicznym gdzie okazami są typy danych!



* Żeby zrozumieć, czym są typy danych w Pythonie, wyobraźmy sobie, że mamy ogród zoologiczny. 
* W tym zoo mamy różne rodzaje zwierząt, a każda grupa zwierząt ma swoje specyficzne cechy i zachowania.
* Podobnie w Pythonie mamy różne typy danych, a każdy z nich ma swoje specyficzne właściwości i zachowanie.


## Liczby całkowite (Integer)

```python
liczba_cukierkow = 5
```

## Liczby zmiennoprzecinkowe (Float)

```python
ilosc_wody_litry = 2.5
```

## Łańcuchy znaków (String) `str`

```python
imie = "Andrzej"
```

## Listy (List)
Wyobraźmy sobie, że mamy listę wszystkich zwierząt w zoo. Lista może zawierać różne typy zwierząt: "Lew", "Tygrys", "Słoń".

W Pythonie, takie zestawienia to listy, czyli typ danych `list`.


```python
lista_zwierzat = ["Zebra", "Słon", "Gepard"]
```

## Słowniki (Dictionary)
Z pewnością w zoo mamy również szczegółowe informacje o każdym zwierzęciu, np. imię, gatunek, wiek. 

Te szczegółowe informacje możemy przedstawić w formie słownika.

W Pythonie, taki typ danych to słownik, czyli `dict`.


```python
informacje_zebra = {
    "imie": "Marty",
    "gatunek": "Zebra",
    "wiek": 5
}
```

## Krotki (Tuple)
Wyobraźmy sobie, że mamy stałe parki zwierząt, które zawsze są razem, np. para lwów. W Pythonie, takie zestawienia to krotki, czyli typ danych `tuple`.


```python
para_lwow = ("Simba", "Nala")
```

## Zbiory (Set)
W zoo mamy również unikalne grupy zwierząt, na przykład stado papug, gdzie każda papuga jest inna i nie ma powtarzających się osobników.
W Pythonie, takie kolekcje przechowamy w zbiorach, czyli `set`.


```python
zbiory_papug = {"Ara", "Kakadu", "Amazonka"}

# lub 
zbiory_papug = set(["Ara", "Kakadu", "Amazonka"])
```

## Podsumowanie

- `int` - liczby całkowite (np. 5, 10)
- `float` - liczby zmiennoprzecinkowe (np. 2.5, 3.14)
- `str` - łańcuchy znaków (np. "Leo", "Zebra")
- `list` - listy (np. ["Zebra", "Słon"])
- `dict` - słowniki (np. {"imie": "Marty", "gatunek": "Zebra"})
- `tuple` - krotki (np. ("Simba", "Nala"))
- `set` - zbiory (np. {"Ara", "Kakadu"})
