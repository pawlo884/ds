# Wartości logiczne w Pythonie

## Czyli jak reprezentować PRAWDĘ i FAŁSZ w Pythonie



* Typ danych `boolean` (w skrócie `bool`) jest typem logicznym, który przyjmuje jedną z dwóch wartości: `True` (prawda) lub `False` (fałsz). 
* Jest on używany do reprezentowania wartości logicznych i jest kluczowy w operacjach warunkowych oraz w kontroli przepływu programu.


## Przykłady

### Podstawowe przykłady zmiennych boolean

```python
is_raining = True
is_sunny = False

print(is_raining)  # Wyświetli: True
print(is_sunny)    # Wyświetli: False
```

### Porównania liczbowe

```python
number = 10

is_greater_than_five = number > 5
is_equal_to_ten = number == 10

print(is_greater_than_five)  # Wyświetli: True
print(is_equal_to_ten)       # Wyświetli: True
```

### Użycie w instrukcji warunkowej

```python
is_hungry = True

if is_hungry:
    print("Czas coś zjeść!")  # Ten kod się wykona, gdyż is_hungry jest True
else:
    print("Nie jestem głodny.")
```

### Operatory logiczne

*Istnieją trzy główne operatory logiczne w Pythonie: `and`, `or`, oraz `not`.*

- `and`: Zwraca `True` tylko wtedy, gdy obie strony są `True`
- `or`: Zwraca `True`, gdy przynajmniej jedna strona jest `True`
- `not`: Zwraca przeciwną wartość logiczną

```python
is_weekend = True
is_holiday = False

can_sleep_in = is_weekend or is_holiday
must_go_to_work = not is_weekend and not is_holiday

print(can_sleep_in)    # Wyświetli: True (ponieważ is_weekend jest True)
print(must_go_to_work) # Wyświetli: False (ponieważ is_weekend jest True)
```

## Często pojawiające się błędy

### Literówki w wartościach boolean
Czyli używanie `true` zamiast `True` i `false` zamiast `False`. W Pythonie wartości logiczne muszą być zapisane wielką literą.


```python
is_sunny = true  # Nieprawidłowe, powinno być `True`
```

Poprawione:

```python
is_sunny = True  # Poprawne
```

### Przypisanie wartości logicznej zamiast porównania
Często zdarza się, że zamiast porównania przypadkowo przypiszemy wartości logiczne.


```python
age = 18
is_adult = age = 18  # Nieprawidłowe, przypisanie zamiast porównania
```

Poprawione:

```python
age = 18
is_adult = (age == 18)  # Poprawne porównanie
```

### Użycie operatorów `and`, `or`, `not` zamiast `&&`, `||`, `!` (częste w dla programistów przychodzących z innych języków)
W Pythonie używamy słów kluczowych `and`, `or`, `not` zamiast symboli `&&`, `||`, `!`.


```python
is_sunny = True
has_sunglasses = False
can_go_out = is_sunny && has_sunglasses  # Nieprawidłowe, użycie `&&` zamiast `and`
```

Poprawione:

```python
is_sunny = True
has_sunglasses = False
can_go_out = is_sunny and has_sunglasses  # Poprawne
```

### Zapominanie o nawiasach w złożonych warunkach logicznych
Złożone warunki logiczne mogą wymagać nawiasów w celu określenia kolejności operacji.


```python
is_sunny = True
has_sunscreen = False
feels_like_going_out = True
can_go_out = is_sunny and has_sunscreen or feels_like_going_out  # Może być źle zrozumiane
```

Poprawione:

```python
is_sunny = True
has_sunscreen = False
feels_like_going_out = True
can_go_out = (is_sunny and has_sunscreen) or feels_like_going_out  # Jawnie określony porządek operacji
```
