# Zmienne w Pythonie

## Czyli jak przechowywać dane w pamięci komputera



## Co to jest zmienna?
* W programowaniu, zmienna to miejsce, w którym można przechowywać dane. 
* Możesz sobie to wyobrazić jako pojemnik, w którym przechowujesz różne wartości – jak liczby, tekst, czy wyniki obliczeń.

## Jak działają zmienne w Pythonie?
* W Pythonie zmienne są dynamiczne, co oznacza, że nie trzeba określać ich typu podczas deklaracji. 
* Typ zmiennej określa się na podstawie przypisanej wartości.


## Tworzenie zmiennych
Aby stworzyć zmienną, wystarczy nadać jej **nazwę** i **przypisać wartość** za pomocą znaku równości `=`.

`NAZWA PUDEŁKA = CO W NIM UMIEŚCIĆ`

Oto kilka przykładów:


```python
# Przypisanie liczby do zmiennej
wiek = 25

# Przypisanie tekstu do zmiennej
imie = "Jan"

# Przypisanie wartości dziesiętnej do zmiennej
waga = 70.5

# Przypisanie wartości logicznej do zmiennej
czy_student = True
```

## Przykład z życia codziennego
Wyobraźmy sobie, że chcesz przechowywać informacje o osobie: imię, wiek, wagę i czy jest studentem. Tworzysz zmienne dla każdej z tych informacji:


```python
imie = "Anna"
wiek = 22
waga = 58.4
czy_student = True

# Wyświetlenie zmiennych
print("Imię:", imie)
print("Wiek:", wiek)
print("Waga:", waga)
print("Czy student?", czy_student)
```
