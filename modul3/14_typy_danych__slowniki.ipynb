# Słowniki w Pythonie

## Czyli jak przechowywać wartości pod kluczem ;-) 



* Słownik (ang. dictionary) to jeden z podstawowych typów danych w Pythonie, który pozwala na przechowywanie par klucz-wartość. 
* Możesz myśleć o słownikach jak o książce adresowej, gdzie każdy kontakt (wartość) jest kojarzony z unikalnym adresem (kluczem).

## Przykłady

### Tworzenie słownika

* Słownik tworzymy za pomocą nawiasów klamrowych `{}`.
* Każda para klucz-wartość jest oddzielona przecinkiem.
* Klucz i wartość są rozdzielane dwukropkiem `:`.

```python
# Przykład słownika, który zawiera informacje o osobie
person = {
    "name": "Maciej",
    "age": 39,
}
```

```python
# Przykład słownika, który zawiera informacje o kilku klientach:
klienci = {
    "Anna": {"wiek": 28, "miasto": "Warszawa"},
    "Jan": {"wiek": 34, "miasto": "Kraków"},
    "Maria": {"wiek": 22, "miasto": "Gdańsk"}
}
```

### Dostęp do danych

Aby uzyskać informacje o konkretnym kliencie, możesz użyć jego imienia jako klucza:


```python
print(klienci["Anna"])
```

### Dodawanie i modyfikowanie elementów

Dodajmy nowego klienta do słownika:


```python
klienci["Piotr"] = {"wiek": 30, "miasto": "Wrocław"}
print(klienci)
```


Modyfikacja istniejących informacji - np. zmieńmy miasto zamieszkania Anny:


```python
klienci["Anna"]["miasto"] = "Poznań"
print(klienci["Anna"])
```

### Usuwanie elementów

Aby usunąć element ze słownika, możesz skorzystać z funkcji `del`:


```python
del klienci["Jan"]
print(klienci)
```

### Iterowanie przez słownik

Możesz przechodzić przez wszystkie pary klucz-wartość w słowniku za pomocą pętli `for`:


```python
for imie, dane in klienci.items():
    print(f"Imię: {imie}, Wiek: {dane['wiek']}, Miasto: {dane['miasto']}")
```

## Przykładowe błędy

### Dostęp do nieistniejącego klucza

Często chcemy uzyskać dostęp do wartości w słowniku, używając klucza. Jednak próba dostępu do klucza, który nie istnieje, wywoła błąd `KeyError`.


```python
customer_info = {"name": "Alice", "age": 30}

# Błąd: KeyError: 'address'
address = customer_info["address"]
```


**Jak tego unikać?** Użyj metody `.get`, która zwraca `None` (lub wartość domyślną, którą możemy określić), jeśli klucz nie istnieje.


```python
address = customer_info.get("address", "Brak adresu")
print(address)  # Wydrukuje: Brak adresu
```

### Modyfikacja słownika podczas iteracji

Modyfikowanie słownika podczas jego iterowania może prowadzić do niespodziewanych zachowań.


```python
prices = {'apple': 1.2, 'banana': 0.5, 'cherry': 2.5}

for fruit in prices:
    # Próba modyfikacji słownika wewnątrz pętli
    prices['orange'] = 1.7  # Błąd: RuntimeError: dictionary changed size during iteration
```


**Jak tego unikać?** Możemy iterować po listę kluczy lub elementów, aby bezpiecznie modyfikować słownik.


```python
prices = {'apple': 1.2, 'banana': 0.5, 'cherry': 2.5}

for fruit in list(prices.keys()):
    prices[fruit] += 0.1  # Zwiększamy cenę każdego owocu o 0.1
prices['orange'] = 1.7  # Możemy dodać nowy element po iteracji
print(prices)
```

### Zagnieżdżone słowniki

Czasami możemy mieć słowniki zagnieżdżone w innych słownikach, co może prowadzić do pomyłek przy dostępie do danych.


```python
employees = {
    "Alice": {"age": 30, "department": "HR"},
    "Bob": {"age": 24, "department": "IT"}
}

# Błąd: KeyError: 'HR'
department = employees["Alice"]["HR"]
```


**Jak tego unikać?** Upewnij się, że poprawnie odwołujesz się do klucza wewnętrznego słownika.


```python
department = employees["Alice"]["department"]
print(department)  # Wydrukuje: HR
```

### Usuwanie elementów słownika
Usuwanie elementów słownika, bez upewnienia się czy klucz istnieje, może prowadzić do błędu `KeyError`.


```python
fruit_prices = {'apple': 1.2, 'banana': 0.5, 'cherry': 2.5}

# Błąd: KeyError: 'orange'
del fruit_prices['orange']
```


**Jak tego unikać?** Używanie metody `pop`, która może przyjąć wartość domyślną.


```python
fruit_prices.pop('orange', None)  # Nic nie zrobi, jeśli 'orange' nie istnieje
print(fruit_prices)  # Bezpiecznie usuwa klucze, które istnieją
```
