# Listy w Pythonie

## Czyli jak przechowywać wiele elementów w jednej zmiennej



* Listy są uniwersalnym typem danych w Pythonie, który umożliwia przechowywanie uporządkowanej kolekcji elementów, takich jak liczby, słowa, a nawet inne listy.
* Listy są niezwykle przydatne, gdy chcesz pracować z zestawem danych.


## Co to jest lista?

* Lista to kontener, który może przechowywać wiele elementów. 
* Można je traktować jak szereg skrzynek, w których przechowujemy różne wartości.
* Listy są definiowane za pomocą nawiasów kwadratowych `[]`, a elementy są oddzielane przecinkami `,`.


## Przykłady

### Lista liczb

```python
liczby = [1, 2, 3, 4, 5]
```

### Lista słów (ciągów znaków)

```python
slowa = ["jabłko", "banan", "czereśnia"]
```

### Lista mieszanych typów danych

```python
mieszana_lista = [1, "szczęście", 3.14, True]
```

### Lista list

```python
lista_list = [[1, 2, 3], ["a", "b", "c"], [True, False]]
```

## Przykłady z rzeczywistości

### Lista zakupów

Chcesz zapisać listę produktów do kupienia w sklepie.

```python
lista_zakupow = ["chleb", "mleko", "jajka", "masło"]
```

### Lista uczniów w klasie

Prowadzisz listę imion uczniów w klasie.

```python
lista_uczniow = ["Anna", "Bartek", "Cezary", "Daria"]
```

## Podstawowe operacje na listach

### Dodawanie elementu do listy (append)

```python
lista_zakupow = ["chleb", "mleko", "jajka", "masło"]
lista_zakupow.append("ser")
lista_zakupow
# lista_zakupow teraz wygląda tak: ["chleb", "mleko", "jajka", "masło", "ser"]
```

### Dostęp do elementów listy (indeksowanie)

- Pierwszy element (indeks 0):

```python
pierwszy_produkt = lista_zakupow[0]
pierwszy_produkt
# wynik: "chleb"
```

- Ostatni element (indeks -1):

```python
ostatni_produkt = lista_zakupow[-1]
ostatni_produkt
# wynik: "ser"
```

### Modyfikacja elementu w liście

```python
lista_zakupow[1] = "mleko sojowe"
lista_zakupow
# lista_zakupow teraz wygląda tak: ["chleb", "mleko sojowe", "jajka", "masło", "ser"]
```

### Usuwanie elementu z listy

- Usuwanie konkretnego elementu (remove):

```python
lista_zakupow.remove("masło")
lista_zakupow
# lista_zakupow teraz wygląda tak: ["chleb", "mleko sojowe", "jajka", "ser"]
```

- Usuwanie elementu na określonej pozycji (pop):

```python
lista_zakupow.pop(2)
# lista_zakupow teraz wygląda tak: ["chleb", "mleko sojowe", "ser"]
```

Listy są bardzo elastyczne i dlatego często używane w Pythonie. 
Praca z listami jest podstawą do rozwiązywania wielu problemów programistycznych.

## Przykładowe błędy

### IndexError: list index out of range

Błąd ten występuje, gdy próbujemy uzyskać dostęp do elementu listy za pomocą indeksu, który jest poza zakresem istniejących indeksów.


#### Przykład

```python
numbers = [1, 2, 3]

# Próba dostępu do elementu o indeksie 3, który nie istnieje (indeksy zaczynają się od 0)
print(numbers[3])
```

#### Poprawne użycie

```python
numbers = [1, 2, 3]

# Użycie poprawnego indeksu
print(numbers[2])
```

### TypeError: 'int' object is not iterable

Błąd ten występuje, gdy próbujemy iterować po obiekcie, który nie jest iterowalny.


#### Przykład

```python
# Próbujemy iterować po liczbie, co jest błędne
for num in 123:
    print(num)
```

#### Poprawne użycie

```python
# Iterujemy po liście liczb
numbers = [1, 2, 3]
for num in numbers:
    print(num)
```

### ValueError: list.remove(x): x not in list

Błąd ten występuje, gdy próbujemy usunąć element, który nie istnieje w liście.


#### Przykład

```python
numbers = [1, 2, 3]

# Próba usunięcia elementu 4, który nie istnieje w liście
numbers.remove(4)
```

#### Poprawne użycie

```python
numbers = [1, 2, 3]

# Usunięcie elementu 2, który istnieje w liście
numbers.remove(2)
```

### TypeError: can only concatenate list (not "int") to list

Błąd ten występuje, gdy próbujemy dodać do listy coś, co nią nie jest, bez odpowiedniego opakowania.


#### Przykład

```python
numbers = [1, 2, 3]

# Próba dodania liczby do listy bez jej zamiany na listę
numbers + 4
```

#### Poprawne użycie

```python
numbers = [1, 2, 3]

# Dodanie liczby do listy poprzez opakowanie jej w listę
numbers += [4]
print(numbers)
```
