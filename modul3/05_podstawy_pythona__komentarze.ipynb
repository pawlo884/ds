# Komentarze w Pythonie

## Czyli jak schować swoje myśli w kodzie!



Czym są komentarze:
* Komentarze w Pythonie służą do dodawania notatek lub wyjaśnień w kodzie, które nie są wykonywane przez interpreter. 
* Są bardzo przydatne, gdyż pozwalają zrozumieć, co dany fragment kodu robi, zarówno tobie jak i innym, którzy mogą czytać twój kod w przyszłości.
* Komentarze można również używać do tymczasowego wyłączania fragmentów kodu podczas debugowania.

W Pythonie są dwa rodzaje komentarzy:
- Komentarze jednowierszowe
- Komentarze wielowierszowe


## Komentarze jednowierszowe

Komentarze jednowierszowe zaczynają się od znaku `#` i wszystko, co jest po nim w wierszu, jest traktowane jako komentarz.


```python
# To jest komentarz jednowierszowy
print("Hello, World!")  # Możesz dodać komentarz na końcu wiersza
```

## Komentarze wielowierszowe

* W Pythonie nie ma bezpośrednio wsparcia dla komentarzy wielowierszowych jak w niektórych innych językach programowania, ale można używać potrójnych cudzysłowów `""" ... """` do tego celu. 
* W praktyce jednak taki tekst zostanie potraktowany jako string, który nie jest przypisany do żadnej zmiennej i zostanie zignorowany przez interpreter, dlatego często stosuje się wielokrotne użycie znaku `#`.


```python
"""
To jest komentarz wielowierszowy.
Możesz go użyć do dodania dłuższych wyjaśnień w kodzie.
"""
```

## Kiedy używać komentarzy?

- **Wyjaśnianie kodu**: Dodawaj komentarze, aby inni programiści (lub sam Ty po pewnym czasie) mogli łatwo zrozumieć, co robi twój kod.
- **Wskazówki lub przypomnienia**: Używaj komentarzy, aby dodać wskazówki lub przypomnienia dotyczące części kodu, którą planujesz zmienić lub sprawdzić później.
- **Wyłączanie kodu**: Czasem podczas testowania lub debugowania możesz tymczasowo wyłączyć pewne linie kodu poprzez umieszczenie przed nimi znaku `#`.


## Przykład użycia komentarzy

Wyobraź sobie, że piszesz funkcję, która oblicza pole prostokąta:


```python
# Funkcja do obliczania pola prostokąta
def calculate_area(length, width):
    """
    Ta funkcja oblicza pole prostokąta.
    length: długość prostokąta
    width: szerokość prostokąta
    """
    # Mnożymy długość przez szerokość, aby uzyskać pole
    area = length * width
    return area

# Wywołanie funkcji calculate_area z przykładowymi wartościami
print(calculate_area(5, 3))  # Oczekiwany wynik: 15
```

W tym przykładzie używamy komentarzy jednowierszowych, aby wprowadzić i wyjaśnić funkcję, a także dodajemy potrójne cudzysłowy, aby opisać parametry funkcji i ich znaczenie.
