# Łańcuchy znaków w Pythonie

## Czyli jak Python traktuje słowa i zdania



* Stringi (łańcuchy znaków) to podstawowy typ danych w Pythonie, który przechowuje tekst. 
* W Pythonie stringi są zamknięte w pojedynczych (`'`) lub podwójnych (`"`) cudzysłowach.
* Możemy używać ich zamiennie, co jest bardzo wygodne.
* Przykłady stringów w rzeczywistości mogą być wszędzie: nazwy, adresy, wiadomości tekstowe itp.

## Przykłady

### Tworzenie stringów


```python
# Stringi mogą być zamknięte w pojedynczych cudzysłowach
nazwa = 'Katarzyna'
print(nazwa)

# Albo w podwójnych cudzysłowach
zawod = "Programista"
print(zawod)
```

```python
# możemy też tworzyć stringi wielolinijkowe - używając "
wiele_linii = """
Witaj świecie!

Jak się masz?
"""
print(wiele_linii)
```

```python
# możemy też tworzyć stringi wielolinijkowe - używając '
wiele_linii = '''
Witaj świecie!

Jak się masz?
'''
print(wiele_linii)
```

### Łączenie (konkatenacja) stringów

Możemy łączyć dwa lub więcej stringów w jeden za pomocą operatora `+`.


```python
imie = 'Katarzyna'
nazwisko = 'Kowalska'
pelne_imie = imie + ' ' + nazwisko
print(pelne_imie)  # Output: Katarzyna Kowalska
```

### Indeksowanie i slicingi

Możemy uzyskać dostęp do poszczególnych znaków w stringu za pomocą indeksów, gdzie pierwszy znak ma indeks 0.


```python
miasto = "Warszawa"
pierwsza_litera = miasto[0]
ostatnia_litera = miasto[-1]
print(pierwsza_litera)  # Output: W
print(ostatnia_litera)  # Output: a
```

### Metody stringów

Stringi mają wiele przydatnych metod wbudowanych, które pozwalają na manipulowanie nimi. Oto kilka przykładów:

1. `upper()`: zamienia wszystkie litery na wielkie
2. `lower()`: zamienia wszystkie litery na małe
3. `replace()`: zamienia ciąg znaków na inny ciąg znaków
4. `split()`: dzieli string na listę elementów na podstawie separatora


```python
tekst = "Witaj w świecie programowania w Pythonie"

# Zamiana na wielkie litery
print(tekst.upper())  # Output: WITAJ W ŚWIECIE PROGRAMOWANIA W PYTHONIE

# Zamiana na małe litery
print(tekst.lower())  # Output: witaj w świecie programowania w pythonie

# Zamiana jednego słowa na inne
nowy_tekst = tekst.replace("Pythonie", "Java")
print(nowy_tekst)  # Output: Witaj w świecie programowania w Javie

# Dzielenie tekstu na podstawie spacji
slowa = tekst.split(' ')
print(slowa)  # Output: ['Witaj', 'w', 'świecie', 'programowania', 'w', 'Pythonie']
```

Stringi są niezmienne, co oznacza, że raz utworzony string nie może być zmieniony. Wszystkie operacje na stringach tworzą nowe stringi.

### Nieoczekiwane znaki specjalne

Niepoprawne użycie znaków specjalnych w stringach:


```python
path = "C:\new_folder\test.txt"
# Oczekiwany efekt: "C:
# ew_folder	est.txt"
print(path)
```


Python interpretuje `\n` jako nową linię, a `\t` jako tabulator. Można tego uniknąć używając surowych stringów:


```python
path = r"C:\new_folder\test.txt"
print(path)  # Outputs: C:\new_folder\test.txt
```

### Indeksowanie poza zakresem

Próba dostępu do indeksu poza zakresem stringa:


```python
word = "Python"
letter = word[10]  # IndexError: string index out of range
```


Poprawny sposób to upewnić się, że indeks znajduje się w zakresie lub obsługiwać taki przypadek:


```python
word = "Python"
if len(word) > 10:
    letter = word[10]
else:
    print("Index is out of range.")
```

### Błąd typu danych

Próby użycia niekompatybilnych typów danych z operacjami na stringach:


```python
age = 25
message = "I am " + age + " years old."
# TypeError: can only concatenate str (not "int") to str
```


Rozwiązaniem jest konwersja liczby do stringa:


```python
message = "I am " + str(age) + " years old."
print(message)  # Outputs: I am 25 years old.
```

### `f-strings`

Dla bardziej złożonych operacji tekstowych warto używać nowoczesnych metod:


```python
name = "Alice"
age = 30

# Stary sposób:
message = "My name is " + name + " and I am " + str(age) + " years old."
print(message)  # Outputs: My name is Alice and I am 30 years old.

# Najnowszy sposób z użyciem f-strings:
message = f"My name is {name} and I am {age} years old."
print(message)  # Outputs: My name is Alice and I am 30 years old.
```
