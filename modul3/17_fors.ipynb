# Pętle for w Pythonie

## Czyli jak powtarzać czynności 



* Pętla `for` w Pythonie to sposób na wykonanie bloku kodu wielokrotnie, iterując po elementach kolekcji (takiej jak lista, tuple, string itp.). 
* Pętla `for` działa w ten sposób, że przy każdym przebiegu pętli wykonywany jest blok kodu, a zmienna iteracyjna przyjmuje kolejną wartość z kolekcji.

### Ogólna składnia

```python
for zmienna in kolekcja:
    # blok kodu do wykonania
```

## Przykłady

### Prosta iteracja po liście

Załóżmy, że mamy listę liczb i chcemy wypisać każdą z tych liczb:


````python
numbers = [1, 2, 3, 4, 5]

for number in numbers:
    print(number ** 2)
````

### Przykład 1: Lista zakupów

Wyobraź sobie, że masz listę zakupów i chcesz wydrukować każdy przedmiot na liście.


````python
shopping_list = ["jabłka", "chleb", "mleko", "masło"]

for item in shopping_list:
    print(item)
````

### Obliczanie pola prostokątów

Załóżmy, że mamy listę par długości boków prostokątów i chcemy obliczyć ich pola.


````python
rectangles = [(2, 3), (4, 5), (6, 7)]

for length, width in rectangles:
    area = length * width
    print(f'Pole prostokąta o bokach {length} x {width} wynosi {area}')
````

### Słownik z informacjami o klientach

Stwórzmy słownik zawierający informacje o kilku klientach i wypiszmy te informacje.


````python
clients = {
    "Anna": {"wiek": 29, "miasto": "Warszawa"},
    "Jan": {"wiek": 34, "miasto": "Kraków"},
    "Ewa": {"wiek": 25, "miasto": "Gdańsk"}
}

for name, info in clients.items():
    print(f'Klient: {name}')
    print(f'Wiek: {info["wiek"]}')
    print(f'Miasto: {info["miasto"]}\n')
````

### Przykłady błędów

### Błąd z nieprawidłowym zakresem `range()`

**Błąd:**

````python
for i in range(5, 1):
    print(i)
````


**Wyjaśnienie:**
W pętli `for`, funkcja `range()` generuje liczby od startu (tu: 5) do końca (tu: 1), ale wyłączając koniec. Jeśli początkowa liczba jest większa od liczby końcowej, to nic nie zostanie wydrukowane.

**Poprawione:**
Jeśli chcesz wydrukować liczby w odwrotnej kolejności, musisz podać krok:

````python
for i in range(5, 1, -1):
    print(i)
````

### Niezmieniona wartość indeksu w pętli

**Błąd:**

````python
names = ["Anna", "Bob", "Charlie"]
for name in names:
    name.upper()
print(names)
````


**Wyjaśnienie:**
Wartość `i` w pętli `for` jest zmienną tymczasową, i zmiana jej nie modyfikuje oryginalnej listy `names`.

**Poprawione:**
Aby zmodyfikować oryginalną listę, zazwyczaj posługujemy się indeksem:

````python
for i in range(len(names)):
    names[i] = names[i].upper()
print(names)
````

### Błąd typu `ValueError`

**Błąd:**

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}
for key, value in my_dict:
    print(key, value)
````


**Wyjaśnienie:**
Bez użycia metody `.items()`, pętla `for` iteruje tylko po kluczach słownika, co prowadzi do błędu, gdy próbujemy rozpakować dwa elementy (`key, value`).

**Poprawione:**
Użyj `.items()` aby iterować po kluczach i wartościach jednocześnie:

````python
for key, value in my_dict.items():
    print(key, value)
````

### Błąd `IndexError`

**Błąd:**

````python
numbers = [1, 2, 3, 4, 5]
for i in range(6):
    print(numbers[i])
````


**Wyjaśnienie:**
Listy w Pythonie są indeksowane od 0, więc korzystanie z `range(6)` wprowadza próbę dostępu do indeksu 5, który nie istnieje dla pięcioelementowej listy.

**Poprawione:**
Użyj `range(len(numbers))` aby odpowiednio dostosować zakres:

````python
for i in range(len(numbers)):
    print(numbers[i])
````

### Zapomnienie o dwukropkach `:`

**Błąd:**

````python
numbers = [1, 2, 3]
for num in numbers
    print(num)
````


**Wyjaśnienie:**
Python wymaga dwukropka (`:`) na końcu lini, która definiuje blok, jak np. pętla `for`.

**Poprawione:**
Dodaj dwukropek:

````python
for num in numbers:
    print(num)
````
