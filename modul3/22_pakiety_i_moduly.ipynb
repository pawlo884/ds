# Moduły i pakiety w Pythonie

## Czyli jak się organizować



* W Pythonie ważne pojęcia to **pakiety (packages)** i **moduły (modules)**.
* Pomagają one organizować kod w sposób czytelny, ułatwiając zarządzanie projektem i ponowne używanie kodu.

## Moduły (Modules)

* **Moduł** to po prostu plik Pythona (`.py`) zawierający kod: funkcje, klasy, zmienne etc. 
* Możemy go zaimportować do innego pliku Pythona i używać zawartych w nim elementów.

### Tworzenie i używanie modułu 

Stwórzmy plik `math_operations.py`:    

```python    
# math_operations.py    
def add(a, b):
    return a + b
    
def subtract(a, b):
    return a - b    
```

Teraz możemy zaimportować ten moduł do innego pliku, np. 
`main.py`:    
```python    
# main.py
import math_operations    

sum_result = math_operations.add(10, 5)    
print(f'Sum: {sum_result}')  # Sum: 15    
diff_result = math_operations.subtract(10, 5)    
print(f'Difference: {diff_result}')  # Difference: 5    
```

## Pakiety (Packages)

* **Pakiet** to kolekcja modułów grupowanych razem w jednym katalogu.
* Każdy pakiet musi zawierać plik `__init__.py`, choćby pusty, aby Python mógł go rozpoznać jako pakiet.
* Plik `__init__.py` może jednak także zawierać kod inicjalizacyjny dla pakietu.

### Tworzenie i używanie pakietu

Stwórzmy strukturę katalogów dla pakietu:    

```
my_package/
    __init__.py
    geometry.py
    arithmetic.py
```

W pliku `geometry.py`:    
```python
# geometry.py
import math

def area_of_rectangle(width, height):
    return width * height    
    
def area_of_circle(radius):
    return math.pi * (radius ** 2)    

```

W pliku `arithmetic.py`:    
```python    
# arithmetic.py
def add(a, b):        
    return a + b    
    
def subtract(a, b):        
    return a - b    
```

Teraz możemy użyć pakietu `my_package` w innym pliku, np. `main.py`:    

```python    
# main.py    
from my_package.geometry import area_of_rectangle
from my_package.arithmetic import add    

rectangle_area = area_of_rectangle(5, 3)    
print(f'Rectangle Area: {rectangle_area}')  # Rectangle Area: 15    
sum_result = add(10, 5)    
print(f'Sum: {sum_result}')  # Sum: 15    
```

## Przykłady błędów

### Błąd `ModuleNotFoundError`

Jest to jeden z najczęstszych błędów, gdy próbujemy zaimportować moduł, który nie jest zainstalowany.

````python
import non_existent_module
````

### Błąd `ImportError`

Może wystąpić, gdy moduł istnieje, ale ścieżka do niego jest nieprawidłowa albo próbujemy zaimportować coś, co nie istnieje wewnątrz modułu.

````python
# Załóżmy, że mamy plik my_module.py z funkcją my_function
# import inny element, który nie istnieje w my_modulefrom my_module import non_existent_function
from math import ciastko
````

### Błąd `SyntaxError`

Importowanie modułów powinno być jedną z pierwszych rzeczy w pliku Python. Jeśli coś umieścimy przed importem, może to spowodować `SyntaxError`.

````python
print("Hello, World!")import math
````

### Niezgodności nazw

Często zdarza się, że nazwa modułu może kolidować z nazwą zmiennej.

````python
json = {'key': 'value'}  # Przypisanie zmiennej o nazwie 'json' import json  # Próba zaimportowania modułu json# Wynik: 'dict' object has no attribute 'loads'
````

### Błąd `Relative Import`

Stosowanie importów względnych w nieprawidłowy sposób może prowadzić do błędów.

````python
# Załóżmy, że mamy strukturę projektów jak poniżej:
# project/
# ├── main.py
# └── module/
#     └── my_module.py
# Gdy w main.py próbujemy zaimportować my_module:
from .module import my_module  # Błąd w kodzie uruchamianym bezpośrednio# Powinno być:from module import my_module
````
