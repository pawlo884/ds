# Czym jest slicing?

* Slicing to operacja, która pozwala na wybieranie / wycinanie fragmentów z m.in. list i łańcuchów znaków w Pythonie. 
* W skrócie, slicing pozwala na wybieranie fragmentów sekwencji, które są zdefiniowane przez ich indeksy.

zacznijmy od zdefiniowania listy i łańcucha znaków, których będziemy używać w przykładach

```python
a = ["pn", "wt", "śr", "czw", "pt", "sb", "nd"]
a
```

```python
b = "ABCDEFG"
b
```

## Jak pobierać pojedyncze elementy z listy / łańcucha znaków w Pythonie?

```python
# pobieramy pierwszy element listy
a[0]
```

```python
# pobieramy czwarty element listy
a[3]
```

```python
# pobieramy pierwszy element łancucha znaków
b[0]
```

```python
# pobieramy trzeci element łancucha znaków
b[2]
```

## Slicing - pobieranie fragmentów listy / łańcucha znaków w Pythonie "od przodu"



```python
a[0:2]
```

```python
b[0:2]
```

```python
a[0:3]
```

```python
b[0:3]
```

```python
a[:3]
```

```python
b[:3]
```

## Slicing - pobieranie fragmentów listy / łańcucha znaków w Pythonie "od tyłu"



```python
a[4:7]
```

```python
b[4:7]
```

```python
a[4:]
```

```python
b[4:]
```

```python
a[-3:]
```

```python
b[-3:]
```

```python
a[-2:]
```

```python
b[-2:]
```

## Slicing - pobieranie fragmentów listy / łańcucha znaków w Pythonie "ze środka"



```python
a[3:6]
```

```python
b[3:6]
```

```python
a[3:-1]
```

```python
b[3:-1]
```
