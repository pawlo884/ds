# Liczby w Pythonie

## Czyli jak Python traktuje liczby



W Pythonie liczby są jednym z podstawowych typów danych. Python obsługuje kilka różnych typów liczb:

1. **Liczby całkowite (int - integer)** - są to liczby, które nie mają części dziesiętnych, np. 1, -3, 42.

2. **Liczby zmiennoprzecinkowe (float)** - są to liczby, które mają część dziesiętną, np. 3.14, -0.001, 123.456.

## Przykłady

### Dodawanie liczby do siebie

Wyobraźmy sobie, że chcemy obliczyć łączną sumę pieniędzy z naszych dwóch skarbonek.

```python
skarbonka1 = 50  # Zawiera 50 zł
skarbonka2 = 30  # Zawiera 30 zł

suma = skarbonka1 + skarbonka2
print(suma)  # Wynik: 80
```

### Obliczanie pola prostokąta

Załóżmy, że chcemy obliczyć pole prostokąta o szerokości 5 jednostek i długości 10 jednostek.

```python
szerokosc = 5
dlugosc = 10

pole = szerokosc * dlugosc
print(pole)  # Wynik: 50
```

### Uśrednianie wyników z egzaminów

Chcemy obliczyć średnią z trzech ocen: 85.5, 90.0 i 78.5.

```python
ocena1 = 85.5
ocena2 = 90.0
ocena3 = 78.5

srednia = (ocena1 + ocena2 + ocena3) / 3
print(srednia)  # Wynik: 84.66666666666667
```

## Często pojawiające się błędy

### Dzielenie przez zero
Dzielenie przez zero jest operacją matematycznie nieokreśloną i Python zgłasza w takiej sytuacji błąd `ZeroDivisionError`.

```python
# Spróbujmy podzielić przez zero
x = 1
wynik = 10 / (1 - x)
# ZeroDivisionError: division by zero
```

### Nieprawidłowe używanie typów danych

Próba wykonania operacji arytmetycznych na niezgodnych typach danych prowadzi do błędu `TypeError`.

```python
liczba = 10
tekst = "5"

# Nie można dodać liczby do tekstu
wynik = liczba + tekst
# TypeError: unsupported operand type(s) for +: 'int' and 'str'
```

### Przekroczenie zakresu liczb

Choć obecnie Python dynamicznie zarządza dużymi liczbami, może się zdarzyć, że operacje na bardzo dużych liczbach skutkują przepełnieniem pamięci lub długim czasem wykonania.

```python
# Bardzo duża liczba
wynik = 10 ** 1000
print(wynik)
```

### Nieprawidłowe zaokrąglanie liczby zmiennoprzecinkowej
Porównania liczb zmiennoprzecinkowych mogą prowadzić do niespodziewanych rezultatów z powodu ich reprezentacji binarnej.

```python
a = 0.1 + 0.2
b = 0.3

# Może się wydawać, że powinno być True, ale nie jest
print(a == b)
# False
print(f"a: {a}, b: {b}")
# a: 0.30000000000000004, b: 0.3
```

### Nieprawidłowe rzutowanie typów danych

Próba rzutowania (konwersji) tekstu, który nie jest liczbą, do typu liczbowego prowadzi do błędu `ValueError`.

```python
tekst = "abc"

# Próba przekształcenia tekstu na liczbę zakończy się błędem
liczba = int(tekst)
# ValueError: invalid literal for int() with base 10: 'abc'
```
