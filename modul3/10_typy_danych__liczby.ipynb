# Liczby w Pythonie

## Czyli jak Python traktuje liczby



W Pythonie liczby s jednym z podstawowych typw danych. Python obsuguje kilka rnych typw liczb:

1. **Liczby cakowite (int - integer)** - s to liczby, ktre nie maj czci dziesitnych, np. 1, -3, 42.

2. **Liczby zmiennoprzecinkowe (float)** - s to liczby, ktre maj cz dziesitn, np. 3.14, -0.001, 123.456.

## Przykady

### Dodawanie liczby do siebie

Wyobramy sobie, e chcemy obliczy czn sum pienidzy z naszych dwch skarbonek.

```python
skarbonka1 = 50  # Zawiera 50 z
skarbonka2 = 30  # Zawiera 30 z

suma = skarbonka1 + skarbonka2
print(suma)  # Wynik: 80
```

### Obliczanie pola prostokta

Zamy, e chcemy obliczy pole prostokta o szerokoci 5 jednostek i dugoci 10 jednostek.

```python
szerokosc = 5
dlugosc = 10

pole = szerokosc * dlugosc
print(pole)  # Wynik: 50
```

### Urednianie wynikw z egzaminw

Chcemy obliczy redni z trzech ocen: 85.5, 90.0 i 78.5.

```python
ocena1 = 85.5
ocena2 = 90.0
ocena3 = 78.5

srednia = (ocena1 + ocena2 + ocena3) / 3
print(srednia)  # Wynik: 84.66666666666667
```

## Czsto pojawiajce si bdy

### Dzielenie przez zero
Dzielenie przez zero jest operacj matematycznie nieokrelon i Python zgasza w takiej sytuacji bd `ZeroDivisionError`.

```python
# Sprbujmy podzieli przez zero
x = 1
wynik = 10 / (1 - x)
# ZeroDivisionError: division by zero
```

### Nieprawidowe uywanie typw danych

Prba wykonania operacji arytmetycznych na niezgodnych typach danych prowadzi do bdu `TypeError`.

```python
liczba = 10
tekst = "5"

# Nie mona doda liczby do tekstu
wynik = liczba + tekst
# TypeError: unsupported operand type(s) for +: 'int' and 'str'
```

### Przekroczenie zakresu liczb

Cho obecnie Python dynamicznie zarzdza duymi liczbami, moe si zdarzy, e operacje na bardzo duych liczbach skutkuj przepenieniem pamici lub dugim czasem wykonania.

```python
# Bardzo dua liczba
wynik = 10 ** 1000
print(wynik)
```

### Nieprawidowe zaokrglanie liczby zmiennoprzecinkowej
Porwnania liczb zmiennoprzecinkowych mog prowadzi do niespodziewanych rezultatw z powodu ich reprezentacji binarnej.

```python
a = 0.1 + 0.2
b = 0.3

# Moe si wydawa, e powinno by True, ale nie jest
print(a == b)
# False
print(f"a: {a}, b: {b}")
# a: 0.30000000000000004, b: 0.3
```

### Nieprawidowe rzutowanie typw danych

Prba rzutowania (konwersji) tekstu, ktry nie jest liczb, do typu liczbowego prowadzi do bdu `ValueError`.

```python
tekst = "abc"

# Prba przeksztacenia tekstu na liczb zakoczy si bdem
liczba = int(tekst)
# ValueError: invalid literal for int() with base 10: 'abc'
```
