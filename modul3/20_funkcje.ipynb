# Funkcje w Pythonie

## Czyli jak tworzyć własne narzędzia wielokrotnego użytku! 



* Funkcje w Pythonie to bloki kodu zaprojektowane do wykonywania określonych zadań.
* Mogą one przyjmować argumenty (dane wejściowe) i zazwyczaj zwracają jakąś wartość (dane wyjściowe).
* Funkcje pomagają organizować kod, czyniąc go bardziej przejrzystym i wielokrotnie wykorzystywanym.

### Definiowanie funkcji

Aby zdefiniować funkcję, używa się słowa kluczowego `def`, po którym następuje nazwa funkcji, nawiasy okrągłe i dwukropek. Oto podstawowa składnia funkcji:


```python
def nazwa_funkcji(argumenty):
    # blok kodu do wykonania
    return wartość
```

## Przykłady

### Funkcja bez argumentów

Stwórzmy funkcję, która po prostu drukuje wiadomość powitalną:


```python
def przywitaj():
    print("Cześć! Jak się masz?")
```


Aby wywołać tę funkcję, wystarczy napisać:


```python
przywitaj()
```

### Funkcja z argumentami

Teraz stwórzmy funkcję, która oblicza pole prostokąta, przyjmując dwie wartości: długość i szerokość:


```python
def pole_prostokata(dlugosc, szerokosc):
    pole = dlugosc * szerokosc
    return pole
```


Wywołując tę funkcję z argumentami, możemy obliczyć pole dowolnego prostokąta:


```python
wynik = pole_prostokata(5, 3)
print(f"Pole prostokąta wynosi: {wynik}")
```

### Funkcja z wartościami domyślnymi

Możemy także zdefiniować wartości domyślne dla argumentów funkcji. Stwórzmy funkcję, która wita użytkownika i ma opcjonalny argument określający imię:


```python
def przywitaj_uzytkownika(imie = "Przyjacielu"):
    print(f"Cześć, {imie}! Jak się masz?")
```


Teraz możemy wywołać tę funkcję z argumentem lub bez:


```python
przywitaj_uzytkownika("Ewa")
przywitaj_uzytkownika()
```

### Funkcja przyjmująca listę

Napiszmy funkcję, która przyjmuje listę liczb i zwraca ich sumę:


```python
def suma_listy(lista_liczb):
    suma = sum(lista_liczb)
    return suma
```


Możemy teraz przekazać listę liczb do tej funkcji:


```python
liczby = [1, 2, 3, 4, 5, 6]
result = suma_listy(liczby)
print(f"Suma liczb w liście wynosi: {result}")
```

## Przykładowe błędy

### Nieprawidłowe wywołanie funkcji (brak argumentów)
Jeśli funkcja wymaga argumentów, a nie przekażemy ich podczas wywołania, Python zwróci błąd.

```python
def pole_prostokata(dlugosc, szerokosc):
    return dlugosc * szerokosc

# Wywołanie bez argumentów
print(pole_prostokata())
```

### Zbyt wiele argumentów
Jeśli podamy więcej argumentów niż funkcja oczekuje, Python również zgłosi błąd.

```python
def pole_prostokata(dlugosc, szerokosc):
    return dlugosc * szerokosc

# Wywołanie ze zbyt wieloma argumentami
print(pole_prostokata(5, 3, 2))
```

### Błąd składniowy
Użycie niewłaściwej składni lub literówki w kodzie funkcji.

```python
def pole_prostokata(dlugosc, szerokosc):
    retrun dlugosc * szerokosc  # literówka w 'return'

# Poprawne wywołanie funkcji
print(pole_prostokata(5, 3))
```

### Niezdefiniowana zmienna
Jeśli użyjemy zmiennej, która nie została zdefiniowana, Python zgłosi błąd.

```python
def pole_prostokata(dlugosc, szerokosc):
    return dlugosc * szeroksc  

# Poprawne wywołanie funkcji
print(pole_prostokata(5, 3))
```

### Zwracanie `None`
Jeśli funkcja nie zwraca jawnie wartości, zwróci `None`. Może to prowadzić do nieoczekiwanych błędów w programie.

```python
def pole_prostokata(dlugosc, szerokosc):
    wynik = dlugosc * szerokosc  # Brak return

# Poprawne wywołanie funkcji
result = pole_prostokata(5, 3)
print(result)
```
