# Operatory w Pythonie

## Czyli jak wykonujemy operacje na zmiennych i warto≈õciach



* Operatory to symbole, kt√≥re wykorzystywane sƒÖ do wykonywania konkretnych operacji na warto≈õciach lub zmiennych.
* Python, jak ka≈ºdy jƒôzyk programowania, posiada kilka r√≥≈ºnych typ√≥w operator√≥w.

## Przyk≈Çady

Pierwszym typem operator√≥w, jakie poznamy sƒÖ operatory arytmetyczne:

- `+` Dodawanie
- `-` Odejmowanie
- `*` Mno≈ºenie
- `/` Dzielenie
- `%` Modulo (reszta z dzielenia)
- `**` Potƒôgowanie
- `//` Dzielenie ca≈Çkowitoliczbowe

```python
# Dodawanie
a = 5
b = 3
print(a + b)  # Wynik: 8

# Odejmowanie
print(a - b)  # Wynik: 2

# Mno≈ºenie
print(a * b)  # Wynik: 15

# Dzielenie
print(a / b)  # Wynik: 1.6666666666666667

# Modulo
print(a % b)  # Wynik: 2 (reszta z dzielenia 5 przez 3 to 2)

# Potƒôgowanie
print(a ** b)  # Wynik: 125 (5 do potƒôgi 3)

# Dzielenie ca≈Çkowitoliczbowe
print(a // b)  # Wynik: 1 (dzielenie bez reszty)
```

### Operatory por√≥wnania
S≈Çu≈ºƒÖ do por√≥wnywania warto≈õci, zwracajƒÖ `True` lub `False`.

- `==` R√≥wne
- `!=` R√≥≈ºne
- `<` Mniejsze ni≈º
- `>` Wiƒôksze ni≈º
- `<=` Mniejsze lub r√≥wne
- `>=` Wiƒôksze lub r√≥wne

Przyk≈Çady:

```python
a = 5
b = 3

# R√≥wne
print(a == b)  # Wynik: False

# R√≥≈ºne
print(a != b)  # Wynik: True

# Mniejsze ni≈º
print(a < b)  # Wynik: False

# Wiƒôksze ni≈º
print(a > b)  # Wynik: True

# Mniejsze lub r√≥wne
print(a <= b)  # Wynik: False

# Wiƒôksze lub r√≥wne
print(a >= b)  # Wynik: True
```

### Operatory przypisania
U≈ºywane do przypisywania warto≈õci do zmiennych.

- `=` Przypisanie
- `+=` Dodawanie i przypisanie
- `-=` Odejmowanie i przypisanie
- `*=` Mno≈ºenie i przypisanie
- `/=` Dzielenie i przypisanie
- `%=` Modulo i przypisanie
- `**=` Potƒôgowanie i przypisanie
- `//=` Dzielenie ca≈Çkowitoliczbowe i przypisanie

Przyk≈Çady:

```python
a = 5

# Dodawanie i przypisanie
a += 3  # R√≥wnowa≈ºne z a = a + 3
print(a)  # Wynik: 8

# Odejmowanie i przypisanie
a -= 2  # R√≥wnowa≈ºne z a = a - 2
print(a)  # Wynik: 6

# Mno≈ºenie i przypisanie
a *= 2  # R√≥wnowa≈ºne z a = a * 2
print(a)  # Wynik: 12

# Dzielenie i przypisanie
a /= 3  # R√≥wnowa≈ºne z a = a / 3
print(a)  # Wynik: 4.0

# Modulo i przypisanie
a %= 3  # R√≥wnowa≈ºne z a = a % 3
print(a)  # Wynik: 1.0

# Potƒôgowanie i przypisanie
a **= 2  # R√≥wnowa≈ºne z a = a ** 2
print(a)  # Wynik: 1.0

# Dzielenie ca≈Çkowitoliczbowe i przypisanie
a //= 1  # R√≥wnowa≈ºne z a = a // 1
print(a)  # Wynik: 1.0
```

### Operatory logiczne
S≈Çu≈ºƒÖ do ≈ÇƒÖczenia warunk√≥w logicznych.

- `and` Logiczne "i"
- `or` Logiczne "lub"
- `not` Logiczne "nie"

Przyk≈Çady:

```python
a = True
b = False

# Logiczne "i"
print(a and b)  # Wynik: False

# Logiczne "lub"
print(a or b)  # Wynik: True

# Logiczne "nie"
print(not a)  # Wynik: False
```

### Operatory to≈ºsamo≈õciowe
S≈Çu≈ºƒÖ do sprawdzania, czy dwie zmienne odnoszƒÖ siƒô do tego samego obiektu.

- `is` To≈ºsamy z
- `is not` Nie to≈ºsamy z

Przyk≈Çady:

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

# To≈ºsamy z
print(a is b)  # Wynik: True
print(a is c)  # Wynik: False

# Nie to≈ºsamy z
print(a is not c)  # Wynik: True
```

### Operatory przynale≈ºno≈õci
S≈Çu≈ºƒÖ do sprawdzania, czy warto≈õƒá znajduje siƒô w innej warto≈õci (np. w li≈õcie).

- `in` Zawiera siƒô
- `not in` Nie zawiera siƒô

Przyk≈Çady:

```python
a = [1, 2, 3]

# Zawiera siƒô
print(2 in a)  # Wynik: True

# Nie zawiera siƒô
print(4 not in a)  # Wynik: True
```

### Dzielimy przez zero
B≈ÇƒÖd `ZeroDivisionError` pojawia siƒô, gdy pr√≥bujemy podzieliƒá liczbƒô przez zero.


```python
# B≈ÇƒÖd: ZeroDivisionError
result = 10 / 0
```


**Wyja≈õnienie:** W matematyce dzielenie przez zero jest niedozwolone. 


### ≈ÅƒÖczenie r√≥≈ºnych typ√≥w danych
B≈ÇƒÖd `TypeError` mo≈ºe siƒô pojawiƒá, gdy pr√≥bujemy u≈ºyƒá operatora na dw√≥ch r√≥≈ºnych typach danych, kt√≥re siƒô nie zgadzajƒÖ.


```python
# B≈ÇƒÖd: TypeError
number = 10
text = "Hello"
result = number + text
```


**Wyja≈õnienie:** Python nie wie, jak zsumowaƒá liczbƒô (`number`) i tekst (`text`). Mo≈ºna to poprawiƒá np. konwertujƒÖc liczbƒô na tekst za pomocƒÖ `str()`.


```python
number = 10
text = "Hello"
result = text + str(number)  # poprawione
```

B≈Çƒôdy logiczne mogƒÖ wystƒôpowaƒá, je≈õli nie jeste≈õmy ostro≈ºni w pisaniu warunk√≥w.

### Zamiana typu danych nieodpowiedniego do operacji arytmetycznych
Pr√≥ba wykonania operacji arytmetycznych na danych, kt√≥re nie sƒÖ liczbami, mo≈ºe r√≥wnie≈º prowadziƒá do `TypeError`.


```python
# B≈ÇƒÖd: TypeError
result = "5" + 10
```
